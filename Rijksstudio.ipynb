```python
api_key = 'BWkXUz9Z'
url = 'https://www.rijksmuseum.nl/api/nl/collection?key=%s&format=json&type=schilderij&ps=100' % api_key
print(url)
```

```python
import sys
import requests

image_urls = []
diff = sys.maxsize
i = 1
prev_len = 0

while (diff != 0):
    first_len = prev_len
    url += '&p=%d' % i
    response = requests.get(url)
    member_data = response.json()
    image_metadata = member_data['artObjects']
    image_data = list(map(lambda x: x['webImage'], image_metadata))
    image_urls += list(map(lambda x: x['url'], image_data))
    second_len = len(image_urls)
    prev_len = second_len
    diff = second_len - first_len
```
