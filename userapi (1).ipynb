```python
for i in range(len(url)):
  print(url[i])
  r2 = requests.get(url[i])
  members_data = r2.json()
  print(json.dumps(members_data, indent=2))
  objects.append(members_data)
```

```python
objects
```

```python
image_objects = list(map(lambda x: x['userSet']['setItems'], objects))
image_urls = []
for image_obj in image_objects:
  urls = list(map(lambda x: x['image']['cdnUrl'], image_obj))
  image_urls += urls
print(image_urls)
print(len(image_urls))
```

```python
new_image_urls = []
for x in image_urls:
    if x not in new_image_urls:
        new_image_urls.append(x)
```

```python
print(len(new_image_urls))
```

```python
import requests
from PIL import Image

for i in range(0, len(new_image_urls)):
    image = Image.open(requests.get(image_urls[i], stream = True).raw)
    image.save('images/img_%d.jpg' % (i + 1))
```

```python
import requests
from PIL import Image

for i in range(0, len(data)):
    image = Image.open(requests.get(data['Image URL'].iloc[i], stream = True).raw)
    image.save('images/img_%d.jpg' % (i + 1))
```

## Making csv with labels for uncropped images

```python
validation_data = []

for i in range(1, 767):
    path = 'images/img_%d.jpg' % i
    new_row = [path, data['Label (Alvaro)'].iloc[i-1]]
    validation_data.append(new_row)

validation_data
```

```python
pd.DataFrame(validation_data).to_csv("uncropped_images_labelled.csv", index=False)
```
